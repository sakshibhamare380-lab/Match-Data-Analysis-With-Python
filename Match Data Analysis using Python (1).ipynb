```python
import matplotlib.pyplot as plt

# Total values in dataset
total_values = df.size

# Total null values
null_values = df.isnull().sum().sum()

# Non-null values
non_null_values = total_values - null_values

# Data for pie chart
labels = ['Non-Null Values', 'Null Values']
sizes = [non_null_values, null_values]

# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Null vs Non-Null Values in Dataset')
plt.show()

```

```python
import matplotlib.pyplot as plt

# Count duplicate rows
duplicate_count = df.duplicated().sum()

# Data for bar chart
labels = ['Duplicate Records']
values = [duplicate_count]

# Plot bar chart
plt.figure(figsize=(5, 4))
plt.bar(labels, values)
plt.title('Duplicate Records Count')
plt.ylabel('Number of Duplicate Rows')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Setup & Theme
plt.style.use('dark_background')
fig, axes = plt.subplots(3, 3, figsize=(24, 18), facecolor='#0f172a')
fig.suptitle('T20 Match Dynamics & Performance Analytics Dashboard', fontsize=30, fontweight='bold', color='#00d4ff')
plt.subplots_adjust(hspace=0.4, wspace=0.3)

# Color Palette
colors = ['#00d4ff', '#ff9f43', '#0abde3', '#ee5253', '#10ac84', '#feca57', '#ff6b6b']

# --- CHART 1: Score Density ---
sns.kdeplot(df['Match_Total'], fill=True, color=colors[0], ax=axes[0, 0], lw=3)
axes[0, 0].set_title('1. Score Density Profile', fontsize=16, color='white')

# --- CHART 2: Pitch Impact ---
sns.boxplot(x='Pitch_Type', y='Match_Total', data=df, palette='viridis', ax=axes[0, 1])
axes[0, 1].set_title('2. Pitch Type vs Match Total', fontsize=16)

# --- CHART 3: Toss Decision (Pie) ---
toss_data = df['Toss_Decision'].value_counts()
axes[0, 2].pie(toss_data, labels=toss_data.index, autopct='%1.1f%%', colors=[colors[1], colors[2]], startangle=90)
axes[0, 2].set_title('3. Toss Strategy Bias', fontsize=16)

# --- CHART 4: Rank vs Tech Index ---
sns.scatterplot(x='Team_A_Ranking', y='Team_A_Tech_Index', data=df, hue='Pitch_Type', palette='spring', ax=axes[1, 0], s=80)
axes[1, 0].invert_xaxis()
axes[1, 0].set_title('4. Ranking vs Technical Skill', fontsize=16)

# --- CHART 5: Venue Analysis ---
venue_avg = df.groupby('Venue')[['Match_Total', 'Avg_T20_Score_Venue']].mean().head(5) # Top 5 for clarity
venue_avg.plot(kind='bar', ax=axes[1, 1], color=[colors[3], colors[4]])
axes[1, 1].set_title('5. Top Venues: Actual vs Hist. Avg', fontsize=16)
axes[1, 1].tick_params(axis='x', rotation=45)

# --- CHART 6: Tournament Stage ---
sns.violinplot(x='Stage', y='Match_Total', data=df, ax=axes[1, 2], palette='Pastel1')
axes[1, 2].set_title('6. Score Volatility by Stage', fontsize=16)

# --- CHART 7: Correlation Heatmap ---
corr = df[['Team_A_Ranking', 'Team_A_Form', 'Team_A_Tech_Index', 'Match_Total']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', ax=axes[2, 0], cbar=False)
axes[2, 0].set_title('7. Factor Correlation Matrix', fontsize=16)

# --- CHART 8: Head-to-Head ---
h2h = df.groupby('Team_A')['HeadToHead_A_Wins'].mean().sort_values(ascending=False).head(7)
h2h.plot(kind='barh', color=colors[5], ax=axes[2, 1])
axes[2, 1].set_title('8. Avg Historical Wins by Team', fontsize=16)

# --- CHART 9: Scoring Momentum ---
df['Date'] = pd.to_datetime(df['Date'])
temp_df = df.sort_values('Date')
sns.lineplot(x='Date', y='Match_Total', data=temp_df, color=colors[6], marker='o', ax=axes[2, 2])
axes[2, 2].set_title('9. Tournament Scoring Momentum', fontsize=16)
axes[2, 2].tick_params(axis='x', rotation=45)

# Final formatting
plt.show()
```

```python
venue_stats = df.groupby('Venue')['Match_Total'].agg(['mean', 'max', 'min', 'count']).sort_values(by='mean', ascending=False)
print("Top 5 Run-Friendly Venues:")
print(venue_stats.head(5))
```

```python
high_voltage = df[(df['Team_A_Ranking'] <= 5) & (df['Team_B_Ranking'] <= 5) & (df['Match_Total'] > 180)]
print(f"Total High-Voltage Matches found: {len(high_voltage)}")
```

```python
pitch_toss_matrix = pd.crosstab(df['Pitch_Type'], df['Toss_Decision'])
print("Decision Matrix per Pitch Type:")
print(pitch_toss_matrix)
```

```python
def categorize_team(rank):
    if rank <= 5: return 'Elite'
    elif rank <= 12: return 'Mid-Tier'
    else: return 'Underdog'

df['Team_A_Cat'] = df['Team_A_Ranking'].apply(categorize_team)
cat_analysis = df.groupby('Team_A_Cat')['Match_Total'].mean()
print(cat_analysis)
```

```python
volatility = df.groupby('Venue')['Match_Total'].std().sort_values(ascending=False)
print("Most Unpredictable Venues (High Volatility):")
print(volatility.head(5))
```
